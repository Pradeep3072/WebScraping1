```python
url="https://www.flipkart.com/search?q=t+shirts&sid=clo%2Cash%2Cank%2Cedy&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_1_na_na_na&as-pos=1&as-type=RECENT&suggestionId=t+shirts%7CMen%27s+T-shirts&requestId=7ab065f8-3f52-417a-a8df-f671ee96d3d6&as-backfill=on"
```

```python
headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.62", "Accept-Encoding":"gzip, deflate,UTF-8", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
page=requests.get(url)
page
```

```python
soup1=BeautifulSoup(page.content,"html.parser")
soup1
```

```python
soup2=BeautifulSoup(soup1.prettify(),"html.parser")
soup2
```

```python
models=soup2.find("div",class_="_1YokD2 _3Mn1Gg").find_all("div",class_="_1xHGtK _373qXS")
```

```python
import csv 

header = ['Model', 'Price', 'gender']

with open('fk2scrap.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
```

```python
for each in models:
    model = each.find("div",class_="_2WkVRV").text.strip()
    price = each.find("div",class_="_30jeq3").text.strip()
    gender = each.find("a",class_="IRpwTa").text.strip()
    with open('fk2scrap.csv', 'a+', newline='', encoding='UTF8') as f:
        data = [model,price,gender]
        writer = csv.writer(f)
        writer.writerow(data)
```
